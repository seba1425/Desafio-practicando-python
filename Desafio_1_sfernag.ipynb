### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head(10)
```

#1. Análisis de facturación



```python
ventas1= tienda.Precio.sum()
ventas2= tienda2.Precio.sum()
ventas3= tienda3.Precio.sum()
ventas4= tienda4.Precio.sum()
print(f'Total Ventas Tienda 1: {ventas1}')
print(f'Total Ventas Tienda 2: {ventas2}')
print(f'Total Ventas Tienda 3: {ventas3}')
print(f'Total Ventas Tienda 4: {ventas4}')
maximasVentas= max(ventas1,ventas2,ventas3,ventas4)
print(f'Máxima venta: {maximasVentas}')
```

```python
cantVentas1 = tienda.count().Producto
cantVentas2 = tienda2.count().Producto
cantVentas3 = tienda3.count().Producto
cantVentas4 = tienda4.count().Producto
print(f'Cantidad de ventas Tienda 1: {cantVentas1}')
print(f'Cantidad de ventas Tienda 2: {cantVentas2}')
print(f'Cantidad de ventas Tienda 3: {cantVentas3}')
print(f'Cantidad de ventas Tienda 4: {cantVentas4}')

```

# 2. Ventas por categoría

```python
CantidadVentasCategoria1 = tienda.groupby(['Categoría del Producto']).count().Producto
CantidadVentasCategoria2 = tienda2.groupby(['Categoría del Producto']).count().Producto
CantidadVentasCategoria3 = tienda3.groupby(['Categoría del Producto']).count().Producto
CantidadVentasCategoria4 = tienda4.groupby(['Categoría del Producto']).count().Producto


categoria_maxima_ventas1 = CantidadVentasCategoria1.idxmax()
valor_maxima_ventas1 = CantidadVentasCategoria1.max()

categoria_maxima_ventas2 = CantidadVentasCategoria2.idxmax()
valor_maxima_ventas2 = CantidadVentasCategoria2.max()

categoria_maxima_ventas3 = CantidadVentasCategoria3.idxmax()
valor_maxima_ventas3 = CantidadVentasCategoria3.max()

categoria_maxima_ventas4 = CantidadVentasCategoria4.idxmax()
valor_maxima_ventas4 = CantidadVentasCategoria4.max()

print(f'\nLa categoría con la mayor cantidad de ventas en la Tienda 1 es "{categoria_maxima_ventas1}" con {valor_maxima_ventas1} ventas.')
print(f'\nLa categoría con la mayor cantidad de ventas en la Tienda 2 es "{categoria_maxima_ventas2}" con {valor_maxima_ventas2} ventas.')
print(f'\nLa categoría con la mayor cantidad de ventas en la Tienda 3 es "{categoria_maxima_ventas3}" con {valor_maxima_ventas3} ventas.')
print(f'\nLa categoría con la mayor cantidad de ventas en la Tienda 4 es "{categoria_maxima_ventas4}" con {valor_maxima_ventas4} ventas.')
```

```python
ventasCategoria1 = tienda.groupby(['Categoría del Producto']).sum().Precio
ventasCategoria2 = tienda2.groupby(['Categoría del Producto']).sum().Precio
ventasCategoria3 = tienda3.groupby(['Categoría del Producto']).sum().Precio
ventasCategoria4 = tienda4.groupby(['Categoría del Producto']).sum().Precio

print(f'El total de ventas de la tienda 1 por categoria es {ventasCategoria1}')
print(f'El total de ventas de la tienda 2 por categoria es {ventasCategoria2}')
print(f'El total de ventas de la tienda 3 por categoria es {ventasCategoria3}')
print(f'El total de ventas de la tienda 4 por categoria es {ventasCategoria4}')
```

# 3. Calificación promedio de la tienda


```python
valoracion1 = tienda.Calificación.mean()
valoracion2 = tienda2.Calificación.mean()
valoracion3 = tienda3.Calificación.mean()
valoracion4 = tienda4.Calificación.mean()

print(f'La calificación promedio de la tienda 1 es: {valoracion1}')
print(f'La calificación promedio de la tienda 2 es: {valoracion2}')
print(f'La calificación promedio de la tienda 3 es: {valoracion3}')
print(f'La calificación promedio de la tienda 4 es: {valoracion4}')

```

# 4. Productos más y menos vendidos

```python
cantidadproductos1= tienda.groupby(['Producto']).count().Precio
cantidadproductos2= tienda2.groupby(['Producto']).count().Precio
cantidadproductos3= tienda3.groupby(['Producto']).count().Precio
cantidadproductos4= tienda4.groupby(['Producto']).count().Precio

producto_maximo1 = cantidadproductos1.idxmax()
cantidad_maxima1 = cantidadproductos1.max()

producto_maximo2 = cantidadproductos2.idxmax()
cantidad_maxima2 = cantidadproductos2.max()

producto_maximo3 = cantidadproductos3.idxmax()
cantidad_maxima3 = cantidadproductos3.max()

producto_maximo4 = cantidadproductos4.idxmax()
cantidad_maxima4 = cantidadproductos4.max()

print(f'\nEl porducto más vendido de la Tienda 1 es "{producto_maximo1}" con {cantidad_maxima1} productos.')
print(f'\nEl porducto más vendido de la Tienda 2 es "{producto_maximo2}" con {cantidad_maxima2} productos.')
print(f'\nEl porducto más vendido de la Tienda 3 es "{producto_maximo3}" con {cantidad_maxima3} productos.')
print(f'\nEl porducto más vendido de la Tienda 4 es "{producto_maximo4}" con {cantidad_maxima4} productos.')

producto_minimo1 = cantidadproductos1.idxmin()
cantidad_minima1 = cantidadproductos1.min()

producto_minimo2 = cantidadproductos2.idxmin()
cantidad_minima2 = cantidadproductos2.min()

producto_minimo3 = cantidadproductos3.idxmin()
cantidad_minima3 = cantidadproductos3.min()

producto_minimo4 = cantidadproductos4.idxmin()
cantidad_minima4 = cantidadproductos4.min()

print(f'\nEl porducto menos vendido de la Tienda 1 es "{producto_minimo1}" con {cantidad_minima1} productos.')
print(f'\nEl porducto menos vendido de la Tienda 2 es "{producto_minimo2}" con {cantidad_minima2} productos.')
print(f'\nEl porducto menos vendido de la Tienda 3 es "{producto_minimo3}" con {cantidad_minima3} productos.')
print(f'\nEl porducto menos vendido de la Tienda 4 es "{producto_minimo4}" con {cantidad_minima4} productos.')
```

# 5. Envío promedio por tienda




```python
costoEnvio1 = tienda['Costo de envío'].mean()
costoEnvio2 = tienda2['Costo de envío'].mean()
costoEnvio3 = tienda3['Costo de envío'].mean()
costoEnvio4 = tienda4['Costo de envío'].mean()

print(f'El costo promedio de envío de la Tienda 1 es: {costoEnvio1}')
print(f'El costo promedio de envío de la Tienda 2 es: {costoEnvio2}')
print(f'El costo promedio de envío de la Tienda 3 es: {costoEnvio3}')
print(f'El costo promedio de envío de la Tienda 4 es: {costoEnvio4}')
```

# 6. Gráficos



```python
fig, ventas = plt.subplots()
tiendas= ['Tienda 1', 'Tienda 2','Tienda 3','Tienda 4']

ventas.bar(tiendas,[ventas1,ventas2,ventas3,ventas4])
ventas.set_title('Ventas Totales por tienda', loc = "left", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
#ventas.legend()
ventas.set_ylim(1000000000,1200000000) ## ajuste de escala para que la diferencia se aprecie


fig, ventasporciento = plt.subplots()
sizes = [ventas1,ventas2,ventas3,ventas4]
ventasporciento.pie(sizes, labels=tiendas, autopct='%1.1f%%')
ventasporciento.set_title('Porcentaje del total de ventas según tienda', loc = "left", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})

fig, valoracion = plt.subplots()
valoracion.barh(tiendas,[valoracion1,valoracion2,valoracion3,valoracion4])
valoracion.set_xlim(3.95,4.05) ## ajuste de escala para que la diferencia se aprecie
valoracion.set_title('Valoración de los clientes de cada tienda', loc = "left", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
plt.show()


```

```python
fig, ax = plt.subplots(figsize=(9.2, 8))

categories = CantidadVentasCategoria1.index

data = np.array([
    CantidadVentasCategoria1.values,
    CantidadVentasCategoria2.values,
    CantidadVentasCategoria3.values,
    CantidadVentasCategoria4.values
])

data_cum = data.cumsum(axis=1)
category_colors = plt.colormaps['RdYlGn'](np.linspace(0.15, 0.85, data.shape[1]))

ax.invert_yaxis()
ax.xaxis.set_visible(False)
ax.set_xlim(0, np.sum(data, axis=1).max())

for i, (category_name, color) in enumerate(zip(categories, category_colors)):
  widths = data[:, i]
  starts = data_cum[:, i] - widths
  rects = ax.barh(tiendas, widths, left=starts, height=0.8, label=category_name, color=color)

  r, g, b, _ = color
  text_color = 'white' if r * g * b < 0.8 else 'darkgrey'
  ax.bar_label(rects, label_type='center', color=text_color)

ax.legend(ncols=len(categories), bbox_to_anchor=(0, 1),loc='lower left', fontsize='small')
#plt.title('\nCantidad de ventas por categoría y tienda', loc = "left", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
plt.show()
```
